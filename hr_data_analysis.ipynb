```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_hr = pd.read_csv("hr_data.csv")
```

#distribution of employes across the departments

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df_hr, x="Department")
plt.title("Employee count by department")
plt.show()
```

average salary by department

```python
avg_salary = df_hr.groupby('Department')['MonthlyIncome'].mean()
avg_salary.plot(kind='bar',title='Average Monthly Income by Department')
plt.ylabel('Monthly Income')
plt.show()
```

Gender distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df_hr,x='Gender')
plt.title("Gender Distribution")
plt.show()
```

age and monthly income

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df_hr,x='Age',y='MonthlyIncome')
plt.title("Age Vs Monthly Income")
plt.show()
```

marital status

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df_hr,x='MaritalStatus')
plt.title("Marital Status")
plt.show()
```

job role monthly income

```python
plt.figure(figsize=(20,12))
plt.bar(df_hr['JobRole'], df_hr['MonthlyIncome'])
plt.title("Job Role VS Monthly Income")
plt.xlabel("Job Role")
plt.ylabel("Monthly Income")
plt.show()
```
